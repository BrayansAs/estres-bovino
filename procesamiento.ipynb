```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

```python
# Ruta al archivo Excel
archivo_excel = '..\Data-Time line.xlsx'

# Nombre de la hoja que deseas leer
nombre_hoja = '4173-Karol'
nombre_csv = '4173' 

# Leer solo la hoja específica
df = pd.read_excel(archivo_excel, sheet_name=nombre_hoja, engine='openpyxl')
```

```python
df
```

```python
# Eliminar las columnas 'B' y 'C' en el DataFrame original
df.drop(columns=['H','S.H'], inplace=True)
df['Fecha'] = pd.to_datetime(df['Fecha'],format='%d/%m/%Y')
df
```

```python

# Función para convertir horas en formato timedelta
def horas_a_timedelta(horas):
    horas_enteras = int(horas)
    minutos_totales = (horas - horas_enteras) * 60
    minutos_enteros = int(minutos_totales)
    segundos = round((minutos_totales - minutos_enteros) * 60)
    return timedelta(hours=horas_enteras, minutes=minutos_enteros, seconds=segundos)

# Convertir cada fila de horas a timedelta
df['Tiempo'] = df['H.N'].apply(horas_a_timedelta)
```

```python
df
```

```python
# Agregar las columnas "desde" y "hasta"
df["Desde"] = ''
df["Hasta"] = ''

# Calcular las columnas "desde" y "hasta"
for i in range(len(df)):
    if i == 0:
        df.at[i, "Desde"] = pd.to_timedelta('00:00:00')
    else:
        start = df.at[i-1, "Hasta"]
        df.at[i, "Desde"] = start

    end = df.at[i, "Tiempo"] + df.at[i, "Desde"]
    df.at[i, "Hasta"] = end
```

```python
df
```

```python

hourly_periods = []
# Iterar sobre cada fila del dataframe original
for index, row in df.iterrows():
    fecha = row["Fecha"]
    periodo = row["Class"]
    duracion_num = row["H.N"]
    duracion_timedelta = row["Tiempo"]
    desde_timedelta = row["Desde"]
    hasta_timedelta = row["Hasta"]

    start_hour = int(desde_timedelta.total_seconds() // 3600)
    end_hour = int(hasta_timedelta.total_seconds() // 3600)

    # Calcular el total de segundos de duración
    total_seconds = duracion_timedelta.total_seconds()

    for hour in range(start_hour, end_hour + 1):
        # Calcular el tiempo en segundos dentro de la hora actual
        current_hour_start = timedelta(hours=hour )
        current_hour_end = timedelta(hours=(hour + 1) )

        # Calcular la intersección de tiempo dentro de la hora actual
        period_start = max(desde_timedelta, current_hour_start)
        period_end = min(hasta_timedelta, current_hour_end)
        period_seconds = (period_end - period_start).total_seconds()

        # Añadir al DataFrame el periodo correspondiente a la hora actual
        hourly_periods.append({
            "Fecha": fecha,
            "Hora": hour % 24,
            "Periodo": periodo,
            'H.N': duracion_num, 
            "Tiempo": timedelta(seconds=period_seconds),
            "Desde": period_start,
            "Hasta": period_end ,
        })
```

```python
# Crear un DataFrame a partir de la lista de diccionarios
hourly_periods_df = pd.DataFrame( hourly_periods)
#hourly_periods_df.drop(hourly_periods_df.index[-1], inplace=True)
hourly_periods_df
```

```python
hourly_periods_df.drop(columns=['H.N','Desde','Hasta'], inplace=True)
hourly_periods_df
```

```python

# Ahora agrupa y suma por fecha, hora y period
df_grouped = hourly_periods_df.groupby(['Fecha', 'Hora', 'Periodo'])['Tiempo'].sum().reset_index()

# Pivotar y renombrar columnas si es necesario
df_pivot = df_grouped.pivot_table(index=['Fecha', 'Hora'], columns='Periodo', values='Tiempo', fill_value=pd.Timedelta(0)).reset_index()
df_pivot.columns.name = None  # Para quitar el nombre del índice de columnas

# Define the columns to convert
time_columns = ['period eating', 'period other', 'period resting', 'period rumination']

# Convert each timedelta column to decimal hours
for column in time_columns:
    df_pivot[column] = (df_pivot[column].dt.total_seconds() / 3600).round(3)

# Mostrar el DataFrame resultante
df_pivot
```

```python
cm = pd.read_csv("../Data_clima.csv")
cm
```

```python
# Convertir la columna 'Fecha' de cm al mismo tipo de datos que la columna 'fecha' de hourly_periods_df
cm['fecha'] = pd.to_datetime(cm['fecha'],format="%d/%m/%Y")
```

```python

# Fusionar las tablas utilizando la columna de fecha y la de hora
tabla= pd.merge(df_pivot, cm, left_on=['Fecha', 'Hora'], right_on=['fecha', 'HR'], how='left')
# Eliminar la columna duplicada 'Fecha' ya que no es necesaria
tabla.drop(columns=['fecha','HR'], inplace=True)
```
